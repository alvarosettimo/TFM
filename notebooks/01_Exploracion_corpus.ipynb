```python
# verificacion de abstracts solos
abstracts = df['abstract'].tolist()
ruta_salida = "C:\\Users\\alvar\\OneDrive\\Escritorio\\TFM\\data\\abstracts.txt"
# Separador que se usará entre abstracts
separador = "\n\n========================================\n\n"

with open(ruta_salida, 'w', encoding='utf-8') as f:
        for i, abstract in enumerate(abstracts):
            # Escribir el abstract actual
            f.write(abstract)
            # Escribir el separador (excepto después del último)
            if i < len(abstracts) - 1:
                f.write(separador)
```
